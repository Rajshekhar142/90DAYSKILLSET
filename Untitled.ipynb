```python
import torch
from torchvision import models, transforms
from PIL import Image
import matplotlib.pyplot as plt

#load pretrained faster Rcnn network 
model = models.detection.fasterrcnn_resnet50_fpn(pretrained = True)
model.eval()
```

```python
# load the image
image = Image.open('emotions.jpg');

# apply transformation 
transform = transforms.Compose([
    transforms.ToTensor()
])
# add batch dimensions.
image_tensor = transform(image).unsqueeze(0) 

# make prediction
with torch.no_grad():
    prediction = model(image_tensor)
```

```python
plt.figure(figsize = (10,10))
plt.imshow(image)
for element in range(len(prediction[0]['boxes'])):
    box =  prediction[0]['boxes'][element].numpy()
    plt.gca().add_patch(plt.Rectangle((box[0], box[1]), box[2]-box[0] , box[3] - box[1] , linewidth  = 2 , facecolor = 'none' , edgecolor = 'r'))
plt.show()
```
