```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
#sequential as it tell the alignemnt of neural network is linear and is the simplest one of all to implement.
#Dense as each neuron is connected to other neuron and is a base for ann.
from sklearn.preprocessing import MinMaxScaler
```

```python

df = pd.read_csv('code/day3dset/data.csv');
#to get the brief summary of what data looks like..
df.head()
```

```python
#exploratory data science begins...
df.info()
#for getting non null and datatype of columns in df.
```

```python
df.describe()
#for getting common statistical value on columns of dataframe
```

```python
# data preprocessing 
# remove the null row.
print(len(df.columns))
df.head()
```

```python
X= df.drop(columns = ['diagnosis'])
y = df['diagnosis']
y
```

```python
scaler = MinMaxScaler()
X = scaler.fit_transform(X)
X
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
y = encoder.fit_transform(y)
y
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(X,y,shuffle = True , random_state = 42 , train_size = 0.9)
X_train.shape
```

```python
Model = Sequential(
   [Dense(16 , activation = 'relu' , input_dim = 30),
    Dense(8, activation = 'relu'),
    Dense(1, activation = 'sigmoid')]
)
Model.compile(optimizer = 'adam' , loss = 'binary_crossentropy', metrics = ['accuracy'])
Model.summary()
```

```python
history = Model.fit(X_train , y_train , epochs = 150 , validation_split = 0.2)
```

```python
y_pred = Model.predict(X_test)
y_pred = (y_pred > 0.5)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred);
cm
```

```python
sns.heatmap(cm , annot = True)
plt.show()
```
