```python
from collections import Counter
import matplotlib.pyplot as plt 
friends_no = [10 , 23, 50 , 100 , 72 , 61 , 5 , 3]
friend_counts = Counter(friends_no)
xs = range(101) # largest value is 100
ys = [friend_counts[x] for x in xs] # height is just # of friends
plt.bar(xs, ys)
plt.axis([0, 101, 0, 25])
plt.title("Histogram of Friend Counts")
plt.xlabel("# of friends")
plt.ylabel("# of people")
plt.show()

# using stats
sorted_vals = sorted(friends_no)
# max no. of friends
max_f = max(sorted_vals)
min_f = min(sorted_vals)
print(f'{max_f} and min no.of friends: {min_f}')
```

```python
def mean(x):
    return sum(x) / len(x)
mean(friends_no)
```
