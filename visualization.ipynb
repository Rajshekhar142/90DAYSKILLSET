```python
from matplotlib import pyplot as plt

character_name = ['Toshiro hitsugaya' , 'kenpachi Zaraki' , 'Byakuya kuchiki' , 'Sosuke Aizen' , 'Ichigo Kurosaki' , 'Uliquora']
power_levls = [100 , 230 , 220 , 500 , 500 , 300 ]

plt.plot(character_name , power_levls , marker = 'o' , linestyle = '-' , color = 'green')
plt.title('Power Scalling in Bleach')
plt.xlabel('characters in Bleach')
plt.ylabel('Attacking Power')
plt.xticks(rotation = 45)
plt.show()
```

```python
movies = ['Avatar' , 'Avengers : endgame ' , 'Terminator' , 'Civil War' , 'Krish']
num_oscars = [8 , 12 , 2 , 5 , 15]

x_shift = [i+0.3 for i,_ in enumerate(movies)]
plt.bar(x_shift,num_oscars)
plt.ylabel('Name of Film')
plt.title('Most Oscar winner of All time')
plt.xticks([i for i,_ in enumerate(movies)], movies)
plt.show()
```

```python
party_names = ['Samajwadi Party' , 'BJP' , 'INC' , 'RASHTRIYAY LOK DAL' , 'OTHERS']
seats_gained = [37 , 33  , 6 , 2 , 2]

X_shift = [i+0.3 for i , _ in enumerate(party_names)]
plt.bar(X_shift , seats_gained)

plt.title('Party Wise Result')
plt.ylabel('Seats Gained')

plt.xticks([i + 0.3 for i , _ in enumerate(party_names)] , party_names , rotation = 45)
plt.show()
```

```python
# line charts 
variance = [1, 2, 4, 8, 16, 32, 64, 128, 256]
bias_squared = [256, 128, 64, 32, 16, 8, 4, 2, 1]
total_error = [ x+y for x , y in zip(variance,bias_squared)]

xs = [ i for i , _ in enumerate(variance)]

plt.plot(xs , variance , 'g-' , label = 'variance')
plt.plot(xs , bias_squared , 'r-' , label = 'bias_squared')
plt.plot(xs , total_error , 'b:' , label = 'total_error')

plt.legend(loc = 9)
plt.xlabel('Model Complexity')
plt.title('The Bias-Variance TradeOff')
plt.show()
```

```python
friends = [ 70, 65, 72, 63, 71, 64, 60, 64, 67]
minutes = [175, 170, 205, 120, 220, 130, 105, 145, 190]
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i']
plt.scatter(friends, minutes)
# label each point
for label, friend_count, minute_count in zip(labels, friends, minutes):
 plt.annotate(label,
 xy=(friend_count, minute_count), # put the label with its point
 xytext=(5, -5), # but slightly offset
 textcoords='offset points')
plt.title("Daily Minutes vs. Number of Friends")

plt.xlabel('no. of friends')
plt.ylabel('daily time spent(in minutes)')
plt.show()
```
